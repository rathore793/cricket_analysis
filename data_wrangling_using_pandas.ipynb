```python
#Importing relevant python library
import pandas as pd
```

```python
#Reading CSVs using pandas into dataframes
matches_df = pd.read_csv('data/matches_data.csv')
batting_df = pd.read_csv('data/batting_data.csv')
bowling_df = pd.read_csv('data/bowling_data.csv')
```

```python
matches_df.head()
```

```python
#Dropping the first column of the dataframe
matches_df.drop(columns=matches_df.columns[0], axis=1,  inplace=True)
```

```python
batting_df.head()
```

```python
batting_df.drop(columns=batting_df.columns[0], axis=1,  inplace=True)
```

```python
bowling_df.head()
```

```python
bowling_df.drop(columns=bowling_df.columns[0], axis=1,  inplace=True)
```

```python
#Creating an empty dictionary to store the match ids of all the matches as they will be unique for each match and will be used in connecting with other datas
match_id_dict = {}

for index, row in matches_df.iterrows(): 
    match_id_dict[f"{row['Team1']} Vs {row['Team2']}"] = row['Match_id']
    match_id_dict[f"{row['Team2']} Vs {row['Team1']}"] = row['Match_id']
    
```

```python
# Replacing values of 'U.A.E.' with 'United Arab Emirates' as they may cause errors in future
temp = {}
for key, value in match_id_dict.items():
    if 'U.A.E.' in key:
        n= key.replace('U.A.E.',"United Arab Emirates" )
        temp[key] = n
        
for key, value in temp.items():
    match_id_dict[value] = match_id_dict.pop(key)
```

```python
match_id_dict
```

```python
batting_df.head()
```

```python
# function to check and convert whether the player got Out or not
def out_not_out(string):
    if "not out" in string.lower():
        return "not out"
    else:
        return 'out'
```

```python
#Applying the above function using .apply method
batting_df['out/not_out'] = batting_df['out/not_out'].apply(out_not_out)
```

```python
batting_df.head()
```

```python
batting_df['out/not_out'].unique()
```

```python
#Mapping the values of match_id_dict and creating new column using the same
batting_df['match_id'] = batting_df['match'].map(match_id_dict)
```

```python
bowling_df['match_id'] = bowling_df['match'].map(match_id_dict)
```

```python
# Creating new CSVs file of the updated dataframes
matches_df.to_csv('data/new_matches_data.csv')
batting_df.to_csv('data/new_batting_data.csv')
bowling_df.to_csv('data/new_bowling_data.csv')
```
